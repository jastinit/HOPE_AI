```python
# print Odd Numbers in the list
```

```python
nos = (20,10,16,19,25,1,276,188)
for i in nos:
  if i%2!=0:
    print(f"{i} is odd")
```

```python
# print Even numbers in the list
```

```python
nos = (20,10,16,19,25,1,276,188)
for no in nos:
  if no%2 ==0:
    print(f"{no} is Even")
```
